```python
import scipy.spatial
import numpy as np

f0 = open('xyz' , 'r')
f1 = open('xyzc', 'r', encoding="utf-8")
f2 = open('xyz_v2' , 'r')

# xyz_v2 include more type of pipe
xyz_v2 = []
# xyz include only one type of type
xyz  = []
xyzc = []

class dataset:
    def __init__(self):
        self.pipe_coordinate=list()
        self.connector_coordinate=list()
    def import_pipe(self,pipe):
        self.src_pipe=pipe
    def import_pipe_v2(self,pipe_v2):
        self.src_pipe_v2=pipe_v2
    def import_connector(self,connector):
        self.src_connector=connector
    def pre_processing_pipe(self):
        for i in self.src_pipe:
            tmplist=list()
            for j in i:
                # tmplist.append(j[:3])
                self.pipe_coordinate.append(j[:3])
    def pre_processing_connector(self):
        for i in self.src_connector:
            self.connector_coordinate.append(i[:3])


processor=dataset()

#讀取資料
for i in f0.readlines():
    temp0 = []
    temp1 = []
    temp2 = i.split(', ')

    for j in range(len(temp2)):
        if j <= 2:
            temp0.append(float(temp2[j]))
        elif j == 3:
            temp2[j] = temp2[j].split('\n')
            temp0.append(float(temp2[j][0]))
        elif 3 < j <= 6:
            temp1.append(float(temp2[j]))
        elif j == 7:
            temp2[j] = temp2[j].split('\n')
            temp1.append(float(temp2[j][0]))
    xyz.append([temp0, temp1])

for i in f2.readlines():
    temp0 = []
    temp1 = []
    temp2 = i.split(', ')

    for j in range(len(temp2)):
        if j <= 2:
            temp0.append(float(temp2[j]))
        elif j == 3:
            temp2[j] = temp2[j].split('\n')
            temp0.append(float(temp2[j][0]))
        elif 3 < j <= 6:
            temp1.append(float(temp2[j]))
        elif j == 7:
            temp2[j] = temp2[j].split('\n')
            temp1.append(float(temp2[j][0]))
    xyz_v2.append([temp0, temp1])

for i in f1.readlines():
    temp0 = []
    temp1 = i.split(', ')
    for j in range(0, 5):
        if j <= 2:
            temp0.append(float(temp1[j]))
        elif j == 3:
            temp0.append(temp1[j])
        elif j == 4:
            temp1[j] = temp1[j].split('\n')
            temp0.append(int(temp1[j][0]))
    xyzc.append(temp0)

tmp_src_pipe=list()
tmp_src_pipe_v2=list()

for i in xyz:
    tmp_src_pipe.append(i[0])
    tmp_src_pipe.append(i[1])

for i in xyz_v2:
    tmp_src_pipe_v2.append(i[0])
    tmp_src_pipe_v2.append(i[1])


processor.import_pipe(xyz)
processor.pre_processing_pipe()
processor.import_pipe(tmp_src_pipe)
processor.import_connector(xyzc)
processor.pre_processing_connector()

processor.import_pipe(tmp_src_pipe_v2)
```

```python
distance=scipy.spatial.distance_matrix(
                    processor.pipe_coordinate,
                    processor.connector_coordinate)
```

```python
endpoint_pipe=list()
endpoint_connector=list()
idx=0
for i in distance:
    # suppose all distance > pow(7000,0.5)
    flag=1
    for j in i:
        # if find any distance between pipe and connector is less than pow(7000,0.5
        # It must not a start pipe
        if(j<105):
            flag=0
            break
    if flag:
        endpoint_pipe.append(idx)
        # print(idx)
    idx+=1

# find start and end point of connector
# If distance between a pipe and all connector > pow(7000,0.5)
# then it should be a start connector
# idx=0

for i in range(0,len(distance[0])):
    num_of_next=0
    if(processor.src_connector[i][3]=="M_頂蓋 - 螺紋 - MI - 150 級"):
        num_of_next=1
        continue
    elif(processor.src_connector[i][3]=="M_T 接頭異徑 - 螺紋 - MI - 150 級"):
        num_of_next=3
        continue
    else:
        num_of_next=2
    cnt=0
    for j in range(0,len(distance)):
        # if number of distance between pipe and connector is less than pow(7000,0.5) == 1
        # it must be end of path
        if(distance[j][i]<105):
            cnt+=1
    if cnt<num_of_next:
        endpoint_connector.append([i,cnt])
        # [0]=i
```

```python
connector_next=list()

# all pipe index: 0~2284
visited_pipe=list(range(0,len(processor.pipe_coordinate)))
# print(len(visited_pipe))

# find connector next pipe
for i in range(0,len(distance[0])):

    # case 2: find next for next
    # number of connector next pipe
    num_of_next=0

    if(processor.src_pipe[i][3]=="M_頂蓋 - 螺紋 - MI - 150 級"):
        num_of_next=1
    elif(processor.src_pipe[i][3]=="M_T 接頭異徑 - 螺紋 - MI - 150 級"):
        num_of_next=3
    else:
        num_of_next=2

    # get distance from remaining pipe
    tmp=list()
    for j in range(0,len(visited_pipe)):
        tmp.append(distance[visited_pipe[j]][i])
    tmpidx=list()
    for j in range(0,len(visited_pipe)):   
        tmpidx.append(visited_pipe[j])
    
    # find sorted num according to origin tmp list index
    sort_tmp=sorted(tmp)
    visited_sort_list=list()
    sort_idx=list()
    for j in range(0,len(tmp)): visited_sort_list.append(0)
    for j in range(0,len(tmp)):
        for k in range(0,len(tmp)):
            if(visited_sort_list[k]!=1 and sort_tmp[j]==tmp[k]):
                visited_sort_list[k]=1
                sort_idx.append(tmpidx[k])
                break
    
    # case 1: already endpoint
    for j in endpoint_connector:
        if i==j[0]:
            # print("yes")
            num_of_next=j[1]
            break

    start_idx=0
    list_len=len(sort_idx)
    store_to_connector_next=list()

    if(num_of_next==1):
        store_to_connector_next.append(-1)
        connector_next.append(store_to_connector_next)
        continue

    same_pipe=0
    while(num_of_next and start_idx<list_len):
        # find same if idx/2==idx'/2
        # since we have [0,1], [2,3] as same pipe
        for j in range(0,len(store_to_connector_next)):
            if(int(sort_idx[start_idx]/2)==int(store_to_connector_next[j]/2)):
                same_pipe=1
        if(same_pipe):
            start_idx+=1
            continue
        store_to_connector_next.append(sort_idx[start_idx])
        visited_pipe.remove(sort_idx[start_idx])
        num_of_next-=1
        start_idx+=1
        # print("remove:",sort_idx[start_idx])

    connector_next.append(store_to_connector_next)
    # all conncetor next pipe index
    # print(len(connector_next))
```

```python
import queue
import sys
print("start bfs")

# using list to record idx of pipe and connector we already visited
visited=list()
# all path with idx
paths=list()

visited_connector=list()
for i in range(0,len(distance[0])):
    visited_connector.append(0)
visited_pipe=list()
for i in range(0,len(distance)):
    visited_pipe.append(0)

ans=list()
q=queue.Queue()
start_idx=0
while(1):
    cur=-1
    while(start_idx<len(endpoint_pipe)):
        # print(endpoint_pipe[start_idx])
        if(visited_pipe[endpoint_pipe[start_idx]]==0):
            cur=endpoint_pipe[start_idx]
            break
        start_idx+=1
    # print("cur: ",cur)
    if(cur==-1):break
    q.put(cur)

    pipe_or_connector=1
    tmpans=list()
    
    # print("start again")
    while(q.qsize()>0):
        cur_qsize=q.qsize()
        # print(cur_qsize)
        # print(pipe_or_connector)
        # print(q.qsize())
        for i in range(0,cur_qsize):
            cur_idx=q.get()
            if(pipe_or_connector):
                # pipe
                if visited_pipe[cur_idx]:
                    continue

                # find pipe end
                if cur_idx%2:
                    cur_idx2=cur_idx-1
                else:
                    cur_idx2=cur_idx+1

                # endpoint_pipe.remove(cur_idx)
                tmpans.append([cur_idx,"pipe"])
                tmpans.append([cur_idx2,"pipe"])

                # print("encounter ",cur_idx," id: ",processor.src_pipe[cur_idx][3]," ",cur_idx2," id: ",processor.src_pipe[cur_idx2][3])

                visited_pipe[cur_idx]=1
                visited_pipe[cur_idx2]=1

                # if end is pipe
                if (cur_idx2 in endpoint_pipe):
                    continue
                
                # should append pipe start and end
                # print("pipe ",cur_idx," ",cur_qsize)
                # print(pipe_next[cur_idx2][0])
                q.put(pipe_next[cur_idx2])
                # print("pipe gen",pipe_next[cur_idx2])
                    
            else:
                # connector
                # print("yes")
                if visited_connector[cur_idx]:
                    continue
                visited_connector[cur_idx]=1
                # endpoint_connector.remove(cur_idx)
                tmpans.append([cur_idx,"connector"])

                # print("connector ",cur_idx," ",cur_qsize)
                if(connector_next[cur_idx][0]!=[] and connector_next[cur_idx][0]!=-1):
                    for k in connector_next[cur_idx]:
                        # print("connector produce",k)
                        q.put(k)

        if pipe_or_connector==1:
            pipe_or_connector=0
        else:
            pipe_or_connector=1
        
    ans.append(tmpans)
```
